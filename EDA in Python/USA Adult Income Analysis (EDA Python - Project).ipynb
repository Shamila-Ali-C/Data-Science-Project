### 1. To find Income of Individuals of Different Working CLasses

```python
data.columns
```

#### Plotting graph :

```python
plt.figure(figsize=(12,6))
sns.countplot(x = data['workclass'], hue = data['income'])
plt.title('Income of Individuals of Different Working CLasses',fontweight='bold')
plt.xticks(rotation = 90)
plt.legend()
```

#### Observation :

In the above graph, we have plotted the graph of adults income on the basis of their different working classes.
* Self-Employed is the only category where the number of people who earn more than 50K dollars.
* People working in the Private Sector more than 75% of them earn less than 50K dollars.
* People belonging to the category Without-Pay and Never-Worked not contain any categories.

### 9. To find Adult Income percentage 

#### Changing categorical values to numerical

```python
data.sex.replace({"Female" : 0,"Male" : 1},inplace = True)
```

```python
data.income.replace({"<=50K" : 2,">50K" : 3},inplace = True)
```

#### Doing groupby to draw the pie chart

```python
income_sex = data.groupby(["sex"])["income"].sum()
```

```python
income_sex
```

#### Plotting graph :

```python
income_sex.plot.pie(legend = True,autopct = "%1.2f%%");
```

#### Observation :

In the above chart, we have plotted the graph of adults income percentage on the basis of gender.
* Here 68.86% of adults are male.
* And 31.14% of adults are female.
