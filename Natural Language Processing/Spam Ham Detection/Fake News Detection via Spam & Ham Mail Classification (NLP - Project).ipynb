<h3 style="font-weight: bold; color: darkred;">Overview : </h3>

In the digital communication era, email remains a widely used medium for personal and professional correspondence. However, it is also heavily targeted by spammers, leading to the spread of unwanted content, phishing attempts, and fake information. Differentiating between legitimate (ham) and unsolicited (spam) emails is crucial for maintaining digital hygiene and security.

This project focused on the detection of fake (spam) and legitimate (ham) emails using Natural Language Processing (NLP) and Machine Learning (ML). By leveraging robust text processing pipelines and classification models, we aimed to build a system that can automatically distinguish spam from ham messages with high accuracy.

<h3 style="font-weight: bold; color: darkred;">About the Dataset : </h3>

The dataset used in this project is a labeled collection of email messages, each categorized as either spam (1) or ham (0). It includes both the raw textual content and metadata required for preprocessing and analysis.

<div style="color: white; background-color: #FF0000; font-size: 28px; font-weight: bold; text-align: center; padding: 20px; border-radius: 20px;"> Text Preprocessing</div>

```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    text=re.sub('\w*\d\w*','',text)
    return text
```

```python
spam_ham_data["preprocess_text"] = spam_ham_data["text"].apply(wordopt)
```

```python
spam_ham_data
```

<div style="color: white; background-color: #FF0000; font-size: 28px; font-weight: bold; text-align: center; padding: 20px; border-radius: 20px;"> Lemmatization</div>

```python
lemmatizer = WordNetLemmatizer()
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def lemma_text(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = nltk.word_tokenize(text.lower())
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(lemmatized_words)
```

```python
spam_ham_data['lemmatized_text'] = spam_ham_data['preprocess_text'].apply(lemma_text)
```

```python
spam_ham_data
```

<h3 style="font-weight: bold; color: darkred;">Model Initialization</h3>

```python
models = {
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=200),
    'Decision Tree': DecisionTreeClassifier(),
    'Support Vector Machine': SVC(),
    'Naive Bayes': MultinomialNB(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Bagging': BaggingClassifier(),
    'Extra Trees': ExtraTreesClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

<h3 style="font-weight: bold; color: darkred;">Model Building</h3>

```python
for name, model in models.items():
    print(f"\n Training Model: {name}")
    model.fit(xv_train, y_train)
    y_pred = model.predict(xv_test)

    acc = accuracy_score(y_test, y_pred)
    print(f" Accuracy: {acc:.4f}")
    print(" Classification Report:")
    print(classification_report(y_test, y_pred))
```

<h3 style="font-weight: bold; color: darkred;">Model Evaluation</h3>

```python
results = []

for name, model in models.items():
    print(f"\n Training Model: {name}")
    
    model.fit(xv_train, y_train)
    y_pred = model.predict(xv_test)

    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)

    results.append({
        'Model': name,
        'Accuracy': acc,
        'Precision (1)': report['1']['precision'],
        'Recall (1)': report['1']['recall'],
        'F1-score (1)': report['1']['f1-score'],
        'Weighted Precision': report['weighted avg']['precision'],
        'Weighted Recall': report['weighted avg']['recall'],
        'Weighted F1-score': report['weighted avg']['f1-score']
    })

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.grid(False)
    plt.show()
```
